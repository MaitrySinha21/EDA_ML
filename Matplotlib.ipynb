Scipy:

We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:

1.fitting it to the periodic function
2.plot the fit Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
Max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])


m = np.arange(12)
plt.plot(m, Max)
plt.plot(m, Min)
plt.xlabel('Month')
plt.ylabel('Max & Min Temperature')
plt.show()
```

```
from scipy import optimize
def func(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(func, m, Max, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(func, m, Min, [-40, 20, 0])
```

```
print(res_max)
print(res_min)
```

```
days = np.linspace(0, 12, num=350)

plt.figure()
plt.plot(m, Max,'go')
plt.plot(days, func(days, *res_max))
plt.plot(m, Min,'bo')
plt.plot(days, func(days, *res_min))
plt.xlabel('Month')
plt.ylabel('Temperature')

plt.show()
```

1. Create a pie chart presenting the male/female proportion from titanic data


```
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
```

```
titanic = pd.read_csv(url)
```

```
titanic.head(2)
```

```
s = titanic['sex'].value_counts()
s
```

```
size=(titanic['sex'].value_counts())/len(titanic)*100
fig,ax1 = plt.subplots()
ax1.pie(size,labels=['Male','Female'],autopct='%2.2f%%',shadow=True)
plt.show()
```

2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```
male = titanic[titanic.sex=='male']
female = titanic[titanic.sex=='female']
```

```
plt.figure(figsize=(14,8))
a = male.plot('age','fare',kind='scatter',label='male',color='g',)
female.plot('age','fare',kind='scatter',label='female',color='r',ax=a)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.scatter(male.age,male.fare,c='g',label='male',marker='*')
plt.scatter(female.age,female.fare,c='r',label='female',marker='o')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.show()
```
