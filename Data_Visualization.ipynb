In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```
df=sns.load_dataset('iris')
df.head()
```

```
setosa = df[df.species=='setosa']
virg = df[df.species=='virginica']
vers = df[df.species=='versicolor']
df.species.value_counts()
```

```
x = setosa['sepal_length']
y = setosa['sepal_width']
z = setosa['petal_length']

x1 = virg['sepal_length']
y1 = virg['sepal_width']
z1 = virg['petal_length']

x2 = vers['sepal_length']
y2 = vers['sepal_width']
z2 = vers['petal_length']
```

```
fig = plt.figure(figsize=(14,8))
ax = plt.axes(projection='3d')

ax.scatter(x, y, z,marker='*');
ax.scatter(x1, y1, z1,marker='o');
ax.scatter(x2, y2, z2,marker='^');

ax.set_title('Iris dataset',fontsize=20)
ax.set_xlabel('sepal_length -->')
ax.set_ylabel('sepal_width -->')
ax.set_zlabel('petal_width -->')

plt.show()
```
